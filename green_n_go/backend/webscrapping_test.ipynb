```python
import requests
from datetime import date

meals_time = ["dinner"]
today = date.today()

URL = "https://www.bu.edu/dining/location/marciano/"
page = requests.get(URL)

# print(page.text)
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")
updated_menu = {}
updated_menu[str(today)] = {}
details = ["cals", "saturated fat", "carbs", "sugars", "proteins"]

soup.find()

for meal_time in meals_time:
    id = str(today)+"-"+meal_time
    updated_menu[str(today)][meal_time] = {}
    results = soup.find(id=id)
    xs = results.find_all("li", class_="menu-item menu-main menu-has-warning")
    for x in xs:
        title = x.find("h4", class_="js-nutrition-open-alias menu-item-title").text
        updated_menu[str(today)][meal_time][title] = {}
        updated_menu[str(today)][meal_time][title]["item"] = title
        z = x.find("ul", class_="menu-nutrition-overview")
        if not z==None:
            for line in z:
                line = line.text
                for detail in details:
                    if detail in line:
                        updated_menu[str(today)][meal_time][title][detail]=line
                
    
    
    
#     menus = results.find_all("h4", class_="js-nutrition-open-alias menu-item-title")
    
#     menu_list = [] 
#     for menu in menus:
#         updated_menu[str(today)][meal_time][menu.text]={}
#         menu_list.append(menu.text)
#         updated_menu[str(today)][meal_time][menu.text]["item"]=menu.text
#         xs = results.find_all("li", class_="menu-item menu-main menu-has-warning")
#         for x in xs:
            
```
